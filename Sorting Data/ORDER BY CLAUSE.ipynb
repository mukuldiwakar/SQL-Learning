# ORDER BY Clause

- The ORDER BY clause is used to sort the results of a query.
- You specify the column(s) by which you want to sort the data.

## Ascending order
### This is the default order

```python
%%sql mysql://root:redhat@localhost:3306/mydb

Select * from product order by Sales;
Select * from product order by Sales asc;
# asc order is default order
```

## Descending order
- Arranges the column in descending order.

```python
%%sql mysql://root:redhat@localhost:3306/mydb

Select * from product order by id desc;

```
