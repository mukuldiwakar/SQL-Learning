## Cross Join Overview:

-- A CROSS JOIN is a type of join in SQL that combines every row from one table with every row from another table.
-- It produces a Cartesian product of the two tables, resulting in a potentially large number of rows.



```python
%%sql mysql://root:redhat@localhost:3306/mydb
select * from Cus cross join pro ;
```

```python
%%sql mysql://root:redhat@localhost:3306/mydb
select count(Cus_id) from Cus cross join pro ;
```
